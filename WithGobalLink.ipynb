```python
import google.generativeai as genai
import gradio

genai.configure(api_key="AQ.Ab8RN6L2ebaIjbAqbmvTfEHkmNAwnSdwpKNvzOs_wf9SwNGbmA")  

model = genai.GenerativeModel(
    model_name="gemini-1.5-flash",
    system_instruction="You are a financial expert that specializes in real estate investment and negotiation"
)

chat = model.start_chat(history=[])

def CustomChatGemini(user_input):
    response = chat.send_message(user_input)
    return response.text

demo = gradio.Interface(
    fn=CustomChatGemini,
    inputs="text",
    outputs="text",
    title="My Saving Assistant"
)

demo.launch(share=True)
```
